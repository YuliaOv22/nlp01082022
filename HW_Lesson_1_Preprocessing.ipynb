**1. Удалим @user из всех твитов с помощью паттерна "@[\w]*". Для этого создадим функцию:**
 - **для того, чтобы найти все вхождения паттерна в тексте, необходимо использовать re.findall(pattern, input_txt)**
 - **для для замены @user на пробел, необходимо использовать re.sub()**


```python
df_length = combine_df.tweet.shape[0]
```

```python
def clean_user(data):
  try:
    find_user = re.findall('@[\w]*', data)
    return re.sub(find_user[0], ' ', data)
  except:
    return data  
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(clean_user)
combine_df.head()
```

**2. Изменим регистр твитов на нижний с помощью .lower()**

```python
def get_lower(data):
  return data.lower()
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(get_lower)
combine_df.head()
```

**3. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()), если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).**

```python
def replace_text(data, dict_):
    for word in data.split():        
      if word in dict_.keys():        
        data = re.sub(re.escape(word), dict_[word], data)
        if data.count(word) > 1:
          break
    return data
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(replace_text, dict_=apostrophe_dict)
combine_df.head()
```

**4. Заменим сокращения на их полные формы, используя short_word_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.**

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(replace_text, dict_=short_word_dict)
combine_df.head()
```

**5. Заменим эмотиконы (пример: ":)" = "happy") на пробелы, используя emoticon_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.**

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(replace_text, dict_=emoticon_dict)
combine_df[63:64]
```

**6. Заменим пунктуацию на пробелы, используя re.sub() и паттерн r'[^\w\s]'**

```python
def delete_punctuation(data):
  return re.sub('[^\w\s]', ' ', data)
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(delete_punctuation)
combine_df[10:15]
```

**7. Заменим спец. символы на пробелы, используя re.sub() и паттерн r'[^a-zA-Z0-9]'**

```python
def delete_special_symbols(data):
  return re.sub('[^a-zA-Z0-9]', ' ', data)
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(delete_special_symbols)
combine_df[10:15]
```

**8. Заменим числа на пробелы, используя re.sub() и паттерн r'[^a-zA-Z]'**

```python
def delete_numbers(data):
  return re.sub('[^a-zA-Z]', ' ', data)
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(delete_numbers)
combine_df[10:15]
```

**9. Удалим из текста слова длиной в 1 символ, используя ' '.join([w for w in x.split() if len(w)>1])**

```python
def delete_short_words(data):
  return ' '.join([word for word in data.split() if len(word) > 1])
```

```python
combine_df.tweet = combine_df.tweet[:df_length].apply(delete_short_words)
combine_df[10:15]
```

**10. Поделим твиты на токены с помощью nltk.tokenize.word_tokenize, создав новый столбец 'tweet_token'.**

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def get_tokens(data):
  return nltk.tokenize.word_tokenize(data)
```

```python
combine_df['tweet_token'] = combine_df.tweet[:df_length].apply(get_tokens)
combine_df[10:15]
```

**11. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец 'tweet_token_filtered' без стоп-слов.**

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def delete_stop_words(data):
  stop_words = set(stopwords.words('english'))
  return [word for word in data if not word in stop_words]
```

```python
combine_df['tweet_token_filtered'] = combine_df.tweet_token[:df_length].apply(delete_stop_words)
combine_df[10:15]
```

**12. Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим столбец 'tweet_stemmed' после применения стемминга.**

```python
from nltk.stem import PorterStemmer
```

```python
def stemming_words(data):
  stemmer = PorterStemmer()
  return [stemmer.stem(word) for word in data]
```

```python
combine_df['tweet_stemmed'] = combine_df.tweet_token_filtered[:df_length].apply(stemming_words)
combine_df[10:15]
```

**13. Применим лемматизацию к токенам с помощью nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после применения лемматизации.**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
def lemmating_words(data):
  lemmatizer = WordNetLemmatizer()
  return [lemmatizer.lemmatize(word) for word in data]
```

```python
combine_df['tweet_lemmatized'] = combine_df.tweet_stemmed[:df_length].apply(lemmating_words)
combine_df[25:30]
```
